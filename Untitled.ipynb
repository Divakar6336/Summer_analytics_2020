```python
df_catdumm = pd.get_dummies(df_cat, drop_first=True)
y = df['Attrition']
X = pd.concat([df_catdumm, df_num], axis=1)
```

```python
from sklearn.linear_model import LogisticRegression as LR
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split 
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, roc_curve, auc
from imblearn.over_sampling import SMOTE
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3 , random_state=32)
smk=SMOTE(random_state=42)
#SS = StandardScaler()
# X_train = X_train.select_dtypes(include=['int64'])
# X_train = SS.fit_transform(X_train)
#pca = PCA(n_components=2)
#X_re = pca.fit_transform(X_train)
# X_re = pd.DataFrame(X_re)
# X_final =  pd.concat([df_catdumm, X_re], axis=1)
X_res, y_res = smk.fit_sample(X_train, y_train)
model = LR()
```

```python
X_train.shape
```

```python
y_res.shape
```

```python
model.fit(X_res, y_res)
```

```python
#X_test = pca.transform(X_test)
predict = model.predict(X_test)
probs = model.predict_proba(X_test)
preds = probs[:,1]
fpr, tpr, threshold = roc_curve(y_test, preds)
roc_auc = auc(fpr, tpr)
```

```python
plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

```python
from sklearn.model_selection import StratifiedKFold as SKF
smk1=SMOTE(random_state=42)
model1 = LR()
X_res, y_res = smk1.fit_sample(X, y)
skf = SKF(n_splits=100)
skf.get_n_splits(X_res, y_res)
X_res=X_res.values
y_res = y_res.values
accuracy=[]
for train_index, test_index in skf.split(X_res, y_res):
    #print("TRAIN:", train_index, "TEST:", test_index)
    X_train, X_test = X_res[train_index], X_res[test_index]
    y_train, y_test = y_res[train_index], y_res[test_index]
    model1.fit(X_train, y_train)
    probs = model1.predict_proba(X_test)
    preds = probs[:,1]
    fpr, tpr, threshold = roc_curve(y_test, preds)
    roc_auc = auc(fpr, tpr)
    accuracy.append(roc_auc)
```

```python
print((np.array(accuracy)).mean())
```
